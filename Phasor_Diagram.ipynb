```python
import matplotlib.pyplot as plt
import numpy as np
from math import exp
```

```python
R=8.314
temperatures = np.linspace(1213, 1685, 9)
```

```python
print(temperatures)
```

```python
G_si=[]
G_ge=[]
Xge_s=[]
Xge_l=[]
```

```python
for i in range(len(temperatures)):
  G_si.append(50200-29.8*temperatures[i])
  G_ge.append(36800-30.3*temperatures[i])
  Xge_s.append((1-exp(-G_si[i]/(R*temperatures[i])))/(exp(-G_ge[i]/(R*temperatures[i]))-exp(-G_si[i]/(R*temperatures[i]))))

  rt = R*temperatures[i]
  num = (1-exp((-G_si[i])/rt))*(exp((-G_ge[i])/rt))
  den = (exp((-G_ge[i])/rt) - (exp((-G_si[i])/rt)))
  Xge_l.append(num/den)
```

```python
print(G_si)
print(G_ge)
print(Xge_s)
print(Xge_l)
```

```python
plt.plot(Xge_s,temperatures)
plt.plot(Xge_l,temperatures)
plt.xlabel("XGe")
plt.ylabel("Temperatures(K)")
plt.grid()
plt.text(0.1, 1305, 'SOLID', fontsize = 22)
plt.text(0.4, 1450, 'S+L', fontsize = 22)
plt.text(0.7, 1550, 'LIQUID', fontsize = 22)
plt.legend(["Solidus","Liquidus"])
plt.title("Binary Isomorphous Diagram")
plt.show()
```
